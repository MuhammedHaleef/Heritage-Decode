```python
!pip install pyocr
!pip install pillow
!pip install pyocr[spa,eng]
!pip install googletrans==4.0.0-rc1

from google.colab import files
import io
import pyocr
import pyocr.builders
from PIL import Image
from googletrans import Translator

def translate_to_english(text):
    translator = Translator()
    result = translator.translate(text, src='es', dest='en')
    return result.text

# Display the upload widget.
uploaded = files.upload()

# Check if any files were uploaded.
if not uploaded:
    print('No file was uploaded.')
else:
    # Get the name of the uploaded file.
    filename = list(uploaded.keys())[0]

    # Convert the image data to a PIL image object.
    image_data = uploaded[filename]
    img = Image.open(io.BytesIO(image_data))

    # Initialize OCR tool.
    tools = pyocr.get_available_tools()
    if len(tools) == 0:
        print("No OCR tool found.")
    else:
        tool = tools[0]

        # Use the first available OCR tool to process the image.
        lang = 'spa'  # Set the language code to Spanish
        txt = tool.image_to_string(img, lang=lang, builder=pyocr.builders.TextBuilder())

        # Translate the text from Spanish to English.
        english_translation = translate_to_english(txt)

        print(f"Spanish: {txt}")
        print(f"English: {english_translation}")
```
