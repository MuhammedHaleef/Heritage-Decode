```python
import cv2
import matplotlib.pyplot as plt
```

```python


# Load the image
image = cv2.imread('training/g/g(21).png')

# Check if the image was loaded successfully
if image is not None:
    # Convert the image from BGR to RGB (OpenCV reads images in BGR format)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image using Matplotlib
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis
    plt.title('Inscription g(21)')
    plt.show()
else:
    print("Error: Unable to load the image.")
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.axis('off')  # Hide axis
plt.title('Inscription g(21)')
plt.show()
```

```python
# Edge Detection
```

```python
# Apply the edge detection (Canny detection)
edges = cv2.Canny(gray_image, 100, 200)

threshold_values = 100
_, damaged_mask = cv2.threshold(edges, threshold_values, 255, cv2.THRESH_BINARY)
```

```python
# Inpainting
```

```python
inpainting_radius = 3
restored_image = cv2.inpaint(image, damaged_mask, inpainting_radius, cv2.INPAINT_TELEA)
```

```python
plt.imshow(restored_image)
plt.axis('off')
plt.title('restored_image')
plt.show()
```

```python
# Optionally smooth the inpainted regions
# For example, using Gaussian blur
blurred_image = cv2.GaussianBlur(restored_image, (5, 5), 0)

# Optionally use morphological operations for refinement
# For example, using morphological closing
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
closing = cv2.morphologyEx(blurred_image, cv2.MORPH_CLOSE, kernel)
```

```python

plt.imshow(blurred_image)
plt.axis('off')  # Hide axis
plt.title('Inscription g(21)')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = blurred_image

# Check if the image was loaded successfully
if image is not None:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the image
    _, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)

    # Apply morphological operations to remove small white regions
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)

    # Find contours in the cleaned image
    contours, _ = cv2.findContours(cleaned_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the original image
    result_image = image.copy()
    cv2.drawContours(result_image, contours, -1, (0, 255, 0), 2)

    # Display the result
    plt.imshow(result_image)
    plt.axis('off')
    plt.title('Inscription g(21)')
    plt.show()
else:
    print("Error: Unable to load the image.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = blurred_image

# Check if the image was loaded successfully
if image is not None:
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the image
    _, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)

    # Apply morphological operations to remove small white regions
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
    cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)

    # Find contours in the cleaned image
    contours, _ = cv2.findContours(cleaned_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a mask of the contours
    contour_mask = np.zeros_like(gray_image)
    cv2.drawContours(contour_mask, contours, -1, (255), thickness=cv2.FILLED)

    # Invert the mask
    contour_mask = 255 - contour_mask

    # Apply the mask to the original image
    result_image = cv2.bitwise_and(image, image, mask=contour_mask)

    # Display the result
    plt.imshow(result_image)
    plt.axis('off')
    plt.title('Inscription g(21) without Contours')
    plt.show()
else:
    print("Error: Unable to load the image.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Path to the directory containing the images
image_dir = 'training/g/'

# Get a list of all image filenames in the directory
image_filenames = os.listdir(image_dir)

# Loop through each image file
for filename in image_filenames:
    # Load the image
    image_path = os.path.join(image_dir, filename)
    image = cv2.imread(image_path)

    # Check if the image was loaded successfully
    if image is not None:
        # Apply the same preprocessing steps as before
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        edges = cv2.Canny(gray_image, 100, 200)
        _, damaged_mask = cv2.threshold(edges, 100, 255, cv2.THRESH_BINARY)
        restored_image = cv2.inpaint(image, damaged_mask, 3, cv2.INPAINT_TELEA)
        blurred_image = cv2.GaussianBlur(restored_image, (5, 5), 0)
        gray_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)
        _, binary_image = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)
        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
        cleaned_image = cv2.morphologyEx(binary_image, cv2.MORPH_OPEN, kernel)
        contours, _ = cv2.findContours(cleaned_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        contour_mask = np.zeros_like(gray_image)
        cv2.drawContours(contour_mask, contours, -1, (255), thickness=cv2.FILLED)
        contour_mask = 255 - contour_mask
        result_image = cv2.bitwise_and(image, image, mask=contour_mask)

        # Display the result
        plt.imshow(result_image)
        plt.axis('off')
        plt.title(f'Inscription {filename} without Contours')
        plt.show()
    else:
        print(f"Error: Unable to load the image {filename}.")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Path to the directory containing the images
image_dir = 'training/g/'

# Get a list of all image filenames in the directory
image_filenames = os.listdir(image_dir)

# Load the non-damaged image for comparison
non_damaged_image_path = 'training/g/g(27)_non_damaged.png'
non_damaged_image = cv2.imread(non_damaged_image_path)

# Check if the non-damaged image was loaded successfully
if non_damaged_image is not None:
    # Loop through each image file
    for filename in image_filenames:
        # Load the damaged image
        damaged_image_path = os.path.join(image_dir, filename)
        damaged_image = cv2.imread(damaged_image_path)

        # Check if the damaged image was loaded successfully
        if damaged_image is not None:
            # Resize the damaged image to match the dimensions of the non-damaged image
            damaged_image_resized = cv2.resize(damaged_image, (non_damaged_image.shape[1], non_damaged_image.shape[0]))

            # Compute the absolute difference between the resized damaged and non-damaged images
            diff_image = cv2.absdiff(damaged_image_resized, non_damaged_image)

            # Convert the difference image to grayscale
            gray_diff_image = cv2.cvtColor(diff_image, cv2.COLOR_BGR2GRAY)

            # Threshold the difference image to create a mask of the damaged regions
            _, mask = cv2.threshold(gray_diff_image, 50, 255, cv2.THRESH_BINARY)

            # Inpaint the damaged regions using the mask
            restored_image = cv2.inpaint(damaged_image_resized, mask, 3, cv2.INPAINT_TELEA)

            # Display the original damaged image, the non-damaged image, and the restored image
            fig, axes = plt.subplots(1, 3, figsize=(15, 5))
            axes[0].imshow(cv2.cvtColor(damaged_image_resized, cv2.COLOR_BGR2RGB))
            axes[0].set_title('Damaged Image')
            axes[0].axis('off')
            axes[1].imshow(cv2.cvtColor(non_damaged_image, cv2.COLOR_BGR2RGB))
            axes[1].set_title('Non-Damaged Image')
            axes[1].axis('off')
            axes[2].imshow(cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB))
            axes[2].set_title('Restored Image')
            axes[2].axis('off')
            plt.show()
        else:
            print(f"Error: Unable to load the damaged image for {filename}.")
else:
    print("Error: Unable to load the non-damaged image.")
```
